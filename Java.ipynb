```python
import requests
from bs4 import BeautifulSoup
base_url='https://github.com/topics/Java?page='
final_link=[]
for num in range( 1,8):
    final_link.append(base_url+str(num))
Java_dict = {}
Java_lst = []
SrNo=1
for urls in final_link:
    raw_data = requests.get(urls).text
    soup = BeautifulSoup(raw_data,'lxml')
    all_div=soup.find_all('article', class_="border rounded color-shadow-small color-bg-subtle my-4")
    for item in all_div:
        if(SrNo>200):
            continue
        else:
            repository_name=item.find('a',class_='text-bold wb-break-word').text
            repository_name=repository_name.strip()
            user_name=item.find("h3",class_="f3 color-fg-muted text-normal lh-condensed").text
            user_name=user_name.strip().split("/")[0].strip()
            star=item.find('span',class_='Counter js-social-count').text
            product_link=item.find('a',class_='text-bold wb-break-word')['href']
            product_url = base_url+product_link
            Java_dict = {
                'Sr No':SrNo,
                'User Name':user_name,
                'Repository_Name':repository_name,
                'Star':star,
                'Product URL':product_url
        
                 }
        Java_lst.append(Java_dict)
        SrNo+=1
```

```python
import pandas as pd
final_data= pd.DataFrame(Java_lst)
final_data
```
